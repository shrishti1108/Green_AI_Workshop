```python
df = sns.load_dataset('iris')

# Boxplot
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title("Boxplot of Sepal Length by Species")
plt.show()
```

```python
# Violinplot
sns.violinplot(data=df, x='species', y='sepal_length', palette='muted')
plt.title("Violinplot of Sepal Length by Species")
plt.show()
```

```python
# Heatmap of correlations
numerical_features = df.select_dtypes(include=['number'])
correlation_matrix = numerical_features.corr()  # Calculate the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='species', palette='dark')
plt.suptitle("Pairplot of the Iris Dataset", y=1.02)  # Add a title
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10,7))
sns.scatterplot(x='petal_length',y='petal_width',data=df)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('iris')

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Perform label encoding on 'species' column and overwrite it with numeric labels
df['species'] = label_encoder.fit_transform(df['species'])

# Display the modified dataset
print(df.head())  # Verify the species column now contains numeric labels

# Create a correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap with Encoded Species")
plt.show()
```
